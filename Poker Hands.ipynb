```python
import numpy
import pandas

cards = ['AS','2S','3S','4S','5S','6S','7S','8S','9S','TS','JS','QS','KS',
        'AC','2C','3C','4C','5C','6C','7C','8C','9C','TC','JC','QC','KC',
        'AD','2D','3D','4D','5D','6D','7D','8D','9D','TD','JD','QD','KD',
        'AH','2H','3H','4H','5H','6H','7H','8H','9H','TH','JH','QH','KH']
player_results = pandas.DataFrame(columns=['player','hand','card1','card2','flop1','flop2','flop3','turn','river'])

for i in range(0,1000):
    start_cards = numpy.random.choice(cards,23,replace=False)
    player_pile = start_cards[0:18]
    board_pile = start_cards[18:]
    play1 = numpy.append(player_pile[0:2],board_pile)
    play1 = numpy.insert(play1,0,'P1')
    play1 = numpy.insert(play1,1,i+1)
    play2 = numpy.append(player_pile[2:4],board_pile)
    play2 = numpy.insert(play2,0,'P2')
    play2 = numpy.insert(play2,1,i+1)
    play3 = numpy.append(player_pile[4:6],board_pile)
    play3 = numpy.insert(play3,0,'P3')
    play3 = numpy.insert(play3,1,i+1)
    play4 = numpy.append(player_pile[6:8],board_pile)
    play4 = numpy.insert(play4,0,'P4')
    play4 = numpy.insert(play4,1,i+1)
    play5 = numpy.append(player_pile[8:10],board_pile)
    play5 = numpy.insert(play5,0,'P5')
    play5 = numpy.insert(play5,1,i+1)
    play6 = numpy.append(player_pile[10:12],board_pile)
    play6 = numpy.insert(play6,0,'P6')
    play6 = numpy.insert(play6,1,i+1)
    play7 = numpy.append(player_pile[12:14],board_pile)
    play7 = numpy.insert(play7,0,'P7')
    play7 = numpy.insert(play7,1,i+1)
    play8 = numpy.append(player_pile[14:16],board_pile)
    play8 = numpy.insert(play8,0,'P8')
    play8 = numpy.insert(play8,1,i+1)
    play9 = numpy.append(player_pile[16:18],board_pile)
    play9 = numpy.insert(play9,0,'P9')
    play9 = numpy.insert(play9,1,i+1)
    player_results = player_results.append(dict(zip(player_results.columns, play1)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play2)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play3)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play4)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play5)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play6)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play7)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play8)), ignore_index=True)
    player_results = player_results.append(dict(zip(player_results.columns, play9)), ignore_index=True)
    
full_hand = pandas.DataFrame(player_results['card1'].str[0]+player_results['card2'].str[0]+player_results['flop1'].str[0]+player_results['flop2'].str[0]+player_results['flop3'].str[0]+player_results['turn'].str[0]+player_results['river'].str[0],columns=['hand_numbers'])
full_hand_suits = pandas.DataFrame(player_results['card1'].str[1]+player_results['card2'].str[1]+player_results['flop1'].str[1]+player_results['flop2'].str[1]+player_results['flop3'].str[1]+player_results['turn'].str[1]+player_results['river'].str[1],columns=['hand_suits'])
full_hand['hand_suits'] = pandas.Series(full_hand_suits['hand_suits'])
full_hand['hand_id'] = pandas.Series(player_results['hand'])

for i in list(['A','2','3','4','5','6','7','8','9','T','J','Q','K']):
    full_hand['num'+i] = full_hand['hand_numbers'].str.count(i)

for i in list(['C','S','H','D']):
    full_hand['suit'+i] = full_hand['hand_suits'].str.count(i)
    
full_hand.to_csv('pokerhands.csv')
```
